# Download IEMOCAP Dataset

Downloads the [AbstractTTS/IEMOCAP](https://huggingface.co/datasets/AbstractTTS/IEMOCAP) parquet from HuggingFace (~1.4 GB) and caches it locally.  
Subsequent runs load instantly from cache — no re-download.

```python
from datasets import load_dataset
import pandas as pd
import numpy as np
```

```python
# First run downloads ~1.4 GB and caches locally.
# Every subsequent call reads from cache instantly.
ds = load_dataset("AbstractTTS/IEMOCAP", split="train")
print(f"Downloaded {len(ds):,} utterances")
print(f"Features: {list(ds.features.keys())}")
```

```python
# Convert to DataFrame for inspection
df = ds.to_pandas()
print(df.shape)
df.head(3)
```
